# 1316번 그룹 단어 체커


```python
# 내 풀이
nums = int(input())
cnt = 0
for i in range(nums):
    words = input()
    empty = []
    for j in range(len(words)):
        if words[j] not in empty:
            empty.append(words[j])
        else:
            if empty[-1] == words[j]:
                empty.append(words[j])
            else:
                cnt +=1
                break
print(nums - cnt)
```

### 그냥 단순히 문자열 내에 인덱스 번호만 가지고 비교해도 충분한 문제였음..

```python
# 간단 정답
def g(a): return list(a)==sorted(a,key=a.find)
print(sum(g(input()) for _ in range(int(input()))))

# 쉬운 풀이
N = int(input())
cnt = N

for i in range(N):
    word = input()
    for j in range(0, len(word)-1):
        if word[j] == word[j+1]:
            pass
        elif word[j] in word[j+1:]:
            cnt -= 1
            break

print(cnt)
```
