LV1. 숫자 문자열과 영단어

```python
# 내 풀이
def solution(s):
    s = s.replace('zero','0')
    s = s.replace('one','1')
    s = s.replace('two','2')
    s = s.replace('three','3')
    s = s.replace('four','4')
    s = s.replace('five','5')
    s = s.replace('six','6')
    s = s.replace('seven','7')
    s = s.replace('eight','8')
    s = s.replace('nine','9')
    answer = int(s)
    return answer
```

리스트나 딕셔너리같은 자료형을  enumerates 나 items로 활용할 생각을 하지 못함.  
replace 말고 sub같은 문자열 함수로 적용해도 될듯

```python
# 다른 풀이
num_dic = {"zero":"0", "one":"1", "two":"2", "three":"3", "four":"4", "five":"5", "six":"6", "seven":"7", "eight":"8", "nine":"9"}

def solution(s):
    answer = s
    for key, value in num_dic.items():
        answer = answer.replace(key, value)
    return int(answer)
```
