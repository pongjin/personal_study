### 0322

 1316번 그룹 단어 체커


````python
# 내 풀이
nums = int(input())
cnt = 0
for i in range(nums):
    words = input()
    empty = []
    for j in range(len(words)):
        if words[j] not in empty:
            empty.append(words[j])
        else:
            if empty[-1] == words[j]:
                empty.append(words[j])
            else:
                cnt +=1
                break
print(nums - cnt)
````

***
 10812 : 바구니 순서 바꾸기

````python
# 내 풀이
n,m = map(int,input().split())
ans = list(range(1, n+1))

for i in range(m):
    i, j, k = map(int,input().split())
    i, j, k = i-1, j-1, k-1
    ans = ans[:i] + ans[k:j+1] + ans[i:k] + ans[j+1:]
print(*ans)
````
