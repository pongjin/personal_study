# 2941번 : 크로아티아 알파벳

```python
# 내 풀이
import re
alpha = ['c=','c-','dz=','d-','lj','nj','s=','z=']

word = input()
for i in alpha:
     word = re.sub(i,'1',word)
print(len(word))
```
