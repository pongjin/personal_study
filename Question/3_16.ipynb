1157번 : 단어공부

```python
a = list(input().upper())
b = set(a)
c = [a.count(i) for i in b]
if c.count(max(c)) > 1:
    print('?')
else:
    print(list(b)[c.index(max(c))])
```

정답 : ASCII 코드를 사용해서 인덱싱을 대체함

```python
str = input().upper()
max_num = 0
for i in range(26):
    cnt = str.count(chr(i + 65))
    if max_num < cnt:
        max_num = cnt
        max_char = chr(i + 65)
    elif max_num == cnt:
        max_char = "?"
print(max_char)
```
