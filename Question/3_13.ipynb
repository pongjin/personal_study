2번틀림 : input 을 한번에 해야된다고 생각함

```python
# 정답
while True:
    try:
        a,b = map(int, input().split())
        print(a+b)
    except:
        break
```

sys.stdin.readline() 은 input보다 속도가 빠르다.

```python
# 빠른코딩
import sys

while True:
    try:
        a,b = map(int,sys.stdin.readline().split())
        print(a+b)
    except:
        break
```
